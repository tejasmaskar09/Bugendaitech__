```python
read_sql_query = """
select e.emp_id AS ID, e.enm AS Name, e.salary AS Salary, d.dept AS Department
from emp e
inner join dept d ON e.emp_id = d.emp_id
where e.salary > 50000
order by e.salary

"""

df = pd.read_sql_query(read_sql_query, new)
df
```

```python
read_sql_query = """
select d.dept AS Department, AVG(e.salary) AS Avg_Salary
from emp e
inner JOIN dept d ON e.emp_id = d.emp_id
group by d.dept
having AVG(e.salary) > 50000
"""
df = pd.read_sql_query(read_sql_query, new)
df
```

```python
read_sql_query = """
select e.emp_id AS ID, e.enm AS Name, e.salary AS Salary, d.dept AS Department
from emp e
inner join dept d ON e.emp_id = d.emp_id
where e.salary > (
    SELECT AVG(salary) FROM emp
)
"""
df = pd.read_sql_query(read_sql_query, new)
df
```
