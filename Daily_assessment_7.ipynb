```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dirty_cafe_sales.csv")
df
```

```python
df.isnull().sum()
```

```python
df.isin(['NaN', 'UNKNOWN', 'null']).sum()
```

```python
df.replace(['UNKNOWN', 'ERROR', 'NaN', 'null', 'None'], np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
missing_percent = (df.isna().sum() / len(df)) * 100
print(missing_percent)
```

```python
clean_percent = (df.notna().sum() / len(df)) * 100
print(clean_percent)
```

```python
df
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())

df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean())

df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].median())
```

```python
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df['Transaction Date'] = df['Transaction Date'].fillna(df['Transaction Date'].mode()[0])
```
