```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("rain-agriculture.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])


correlation_matrix = numeric_df.corr()

plt.figure(figsize=(18, 14))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Year", y="RICE YIELD (Kg per ha)", hue="State Name", estimator="mean")
plt.title("Trend of RICE YIELD over the Years by State")
plt.ylabel("RICE YIELD (Kg per ha)")
plt.xlabel("Year")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Average Monthly Rainfall (Monsoon Months)
plt.figure(figsize=(8, 6))
rain_cols = ['JUN', 'JUL', 'AUG', 'SEP']
rain_avg = df[rain_cols].mean()
sns.barplot(x=rain_avg.index, y=rain_avg.values, color="yellowgreen")  
plt.title("Average Monthly Rainfall (Monsoon Months)")
plt.ylabel("Rainfall (mm)")
plt.xlabel("Month")
plt.show()
```

```python
rice_by_year = df.groupby("Year")["RICE PRODUCTION (1000 tons)"].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=rice_by_year.index, y=rice_by_year.values, color='skyblue')
plt.title("Rice Production by Year")
plt.xlabel("Year")
plt.ylabel("Rice Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_years = rice_by_year.sort_values(ascending=False).head(6)

plt.figure(figsize=(8, 8))
plt.pie(top_years.values, labels=top_years.index, autopct='%1.1f%%', startangle=90)
plt.title("Rice Production Share - Top 6 Years")
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x="State Name", y="MAIZE YIELD (Kg per ha)", jitter=True)
plt.title("Maize Yield by State (Strip Plot)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.violinplot(data=df, x="State Name", y="RICE YIELD (Kg per ha)")
plt.title("Distribution of Rice Yield by State (Violin Plot)")
plt.xticks(rotation=90)
plt.savefig("my_plot1.png", dpi=300, bbox_inches='tight') 
plt.tight_layout()
```

```python



plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x="State Name", y="WHEAT YIELD (Kg per ha)")
plt.title("Wheat Yield Distribution by State")
plt.xticks(rotation=90)
plt.savefig("my_plot.png", dpi=300, bbox_inches='tight') 

plt.tight_layout()




```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Year", y="RICE PRODUCTION (1000 tons)", hue="State Name", estimator="mean")
plt.title("Rice Production Trend Over Years")
plt.tight_layout()
plt.savefig("rice_production_trend.png", dpi=300)
plt.show()
```
